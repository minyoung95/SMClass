데이터 조건 검색 (조건문)

```python
# 조건식 생성
filt = df['키']>=185
df[filt]
# df[~filt] : not에 대한 결과값 출력 
df.loc[filt,['국어','수학']]
```

& and연산자, | or연산자

```python
filt = (df['국어']>=70) & (df['키']>190)
df[filt]
```

문자열 함수 : startswith,contains,replace,lower,upper

```python
# startswith() : 해당값으로 시작되는 단어 검색
filt = df['이름'].str.startswith('김')
df[filt]

# contains() : 해당값이 포함되어 있는 단어 검색
filt = df['이름'].str.contains('수')
df.loc[filt, ['이름','학교','키']]

# isin() : 포함이 되어 있는지 확인 // () 괄호 안 리스트타입만 가능
langs = ['Python','Java']
filt = df['SW특기'].isin(langs)
df[filt]

# lower() : 해당값을 모두 소문자로, upper() : 해당값을 모두 대문자로
langs = ['python','java']
filt = df['SW특기'].str.lower().isin(langs)
df[filt]

# 포함이 되어 있으면 모두 검색
filter = df['SW특기'].str.contains('Java',na=False) # NaN값이 있으면 검색이 안됨, na=False로 제거
df[filter]
```

```python
## j가 포함되어 있는 정보 검색 (문자열 함수를 두번쓰면 str도 두번)
filt = df['SW특기'].str.lower().str.contains('j',na=False)
df[filt]
```

```python
filt = df['SW특기'].str.lower().isin(['python','java'])
df[filt]
```

```python
### dropna // axis = 'index'(기본값, row 방향),'columns'(컬럼 방향) , how = 'any'(기본값, 하나라도 있으면 삭제),'all'(전체가 null값일 때 삭제)
df.dropna()

df.dropna(axis='index',how='all') # 모두 NaN row가 없어서 삭제 x
df.dropna(axis='columns',how='all') # 학교 컬럼 삭제
```
