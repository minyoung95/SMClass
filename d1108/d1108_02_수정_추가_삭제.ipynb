```python
import pandas as pd
df = pd.read_excel('score.xlsx',index_col='지원번호')
df
```

데이터 수정

```python
## replace({a:b}) : 데이터 값 a를 b로 변경
df['학교'].replace({'구로고':'가산고'},inplace=True)
df
df['학교'].replace({'가산고':'구로고','디지털고':'단지고'})
```

```python
## 특정 컬럼 값 변경 : 컬럼 = 변경 값 대입
df['SW특기'] = df['SW특기'].str.lower()
df['SW특기'].str.lower() # 소문자
df['SW특기'].str.upper() # 대문자
df['SW특기'].str.capitalize() # 첫글자 대문자

# 컬럼명 = 문자 + 문자
df['학교'] = df['학교']+'등학교'

## 특정 위치(셀) 값 변경
df.loc['1번','국어'] = 100
df.loc['2번',['국어','영어']] = 10,20
df.loc[['2번','3번'],['국어','영어']] = 50
```

데이터 추가

* 컬럼 추가 : df[컬럼명]
* row 추가 : df.loc[index명]

```python
## 데이터 추가 : 컬럼에 없는 컬럼 이름을 입력하면 컬럼 추가
df['합계'] = 0
df['합계'] = df['국어']+df['영어']+df['수학']+df['과학']+df['사회']
df['평균'] = df['합계']/5
df
```

```python
df['결과'] = 'Fail'
## 조건식을 사용하여 값을 변경
filt = df['합계']>= 400
df.loc[filt,'결과'] = 'Pass'
df
```

```python
df['합산'] =100
# ## 국어점수 90 이상, ef['합산']+100
flit = df['국어'] >= 90
df['합산'] = df.loc[flit,'합산']+100
df
```

```python
# 데이터 삭제
df.drop(columns=['합산'],inplace=True)

df.drop(columns=['국어','영어'])
df.drop(columns=['합계','평균','결과']) # 컬럼 삭제
df.drop(columns=df.columns[3:5])
```

```python
# index(row 값) 삭제
df.index
df.drop(index='1번')
df.head(1)
```

```python
# 수작 80점 미만 학생 삭제
filt = df['수학']<80
filt
# df.index : index값 전부 가져오기 // df[filt].index : filt의 조건에 맞는 index 가져오기
df.drop(index=df[filt].index)
```

```python
# 김이 포함되어 있는 학생 삭제
filt = df['이름'].str.contains('김')
df.drop(index=df[filt].index)
```

row 추가

```python
# row 추가 : index 이름 추가 후 list형태로 입력
df.loc['9번'] = ['유승찬','디지털고',184,100,100,100,100,100,'Django',100*5,100/5,'Pass']
df
```

```python
# 9번 디지털고 > 가산고, 'Django' > 'Python'
df.loc['9번',['학교','SW특기']] = ['가산고','Python']
df.loc['7번','이름'] = '김혜정'
df
```
