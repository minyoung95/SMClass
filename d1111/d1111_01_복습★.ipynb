```python
### 문자열 함수
# 함수 사용하여 이름을 첫글자를 제거하여 입력 (str[:] : 문자슬라이싱)
# str.split(",") : ,로 분할, str.startswith('서울') : 서울이 첫 글자인지 확인, str.endswith() : 마지막 글자 확인
# str.contains('s',na=False) : s가 포함된 글자, str.find('x') : x 문자 위치 찾기(몇번째)
# str.replace(',','') : ,를 공백으로 문자 대체
# str.pad(width=20,side='left',fillchar='_') : 문자열 패딩, 특정 글자로 채워줌
# str.zfill(width=10) : 10칸 중 빈칸을 0으로 채워줌, str.strip() : 공백 제거
# str.lower() : 소문자 변경, str.upper() : 대문자 변경
# str.swapcase() : 대소문자 변경
def func(value):
    return value[1:3]
    
df['이름'] = df['이름'].apply(func)
df
```

```python
###
# 1. 요일 추가
# 2. 0~6까지의 숫자를 랜덤으로 입력
# 3. 함수를 사용하여 0:일요일, 1:월요일, ..., 6:토요일

# df['결과'] = '불합격'
# filt = df['평균'] >= 80
# df.loc[filt,'결과'] = '합격'
import random
df['요일'] = 0
for i in range(8):
    df.iloc[i,9] = random.randrange(0,7)

def func(value):
    if value == 0: return '일요일'
    elif value == 1: return '월요일'
    elif value == 2: return '화요일'
    elif value == 3: return '수요일'
    elif value == 4: return '목요일'
    elif value == 5: return '금요일'
    else: return '토요일'

df['요일'] = df['요일'].apply(func)
df
```
