```python
import pandas as pd
import numpy as np

df = pd.read_csv('chipotle.tsv',sep='\t') # tsv파일은 sep='\t', csv파일은 sep=','
df.tail()
```

```python
# 함수 사용하여 item_price $ 제거, float 타입으로 변경

def func(x):
    return float(x[1:])

df['item_price'] = df['item_price'].apply(func)
# df['item_price'].sort_values()
# df.sort_values('item_price')
# df['item_price'].sum()
# df['item_price'].max()
# df['item_price'].min()
```

```python
### lamda 함수 적용
df['item_price'] = df['item_price'].apply(lambda x:float(x[1:]))
```
