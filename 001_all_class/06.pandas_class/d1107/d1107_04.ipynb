```python
df = pd.read_excel('score.xlsx',index_col='이름') # index가 따로 붙음 > row에 있는 데이터를 가져올 때 사용
df
df.sort_index(ascending=False)
```

```python
df.head() # 상단 5개
df.tail() # 하단 5개
df.values # 밸류 값들 
df.shape # 표 크기
df.columns # 컬럼들 나열
df[df.columns[-1]] # 마지막 컬럼의 데이터 확인
df.columns[-1] # 마지막 컬럼 이름 // df[0(번호)]로는 검색이 안된다.
df.info() # 표 데이터 정보

df['키'].min() # 키의 최솟값
df['키'].max() # math의 최댓값
df['키'].mean()
df['키'].nlargest() # 상위 5개 (괄호 안 숫자 입력하면 숫자만큼 출력(기본값 : 5개))
df['키'].nsmallest() # 하위 5개
df['학교'].unique() # 중복제거
df['학교'].nunique() # 중복 갯수
df['SW특기'].count() # null값을 제외 한 갯수
df['학교'].count() # 중복 갯수도 포함
df['국어'].sum() # 합계
```
