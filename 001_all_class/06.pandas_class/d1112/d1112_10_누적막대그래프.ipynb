```python
import pandas as pd
import matplotlib.pyplot as plt
# 한글 설정
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic' # Window용
# matplotlib.rcParams['font.family'] = 'AppleGothic' # Mac 용
matplotlib.rcParams['font.size'] = 15 # 글자크기
matplotlib.rcParams['axes.unicode_minus'] = False # 그래프 마이너스까지 나타낼 수 있게
```

```python
df = pd.read_excel('score.xlsx')
```

누적막대그래프

```python
plt.figure(figsize=(7,4))
plt.bar(df['이름'],df['국어'],label='국어')
plt.bar(df['이름'],df['영어'],bottom=df['국어'],label='영어') # bottom= 국어점수를 기반으로 위쪽으로 그려짐
plt.bar(df['이름'],df['수학'],bottom=df['국어']+df['영어'],label='수학') # bottom= 국어,영어점수를 기반으로 위쪽으로 그려짐
plt.xticks(rotation=45)

for i in range(8):
    plt.text(df['이름'][i],df['국어'][i],df['국어'][i],ha='center')
    plt.text(df['이름'][i],df['국어'][i]+df['영어'][i],df['영어'][i],ha='center')
    plt.text(df['이름'][i],df['국어'][i]+df['영어'][i]+df['수학'][i],df['수학'][i],ha='center')

plt.legend(ncol=3, loc='upper center') # ncol : 가로로 3칸
plt.ylim(0,350)
plt.show()
```
