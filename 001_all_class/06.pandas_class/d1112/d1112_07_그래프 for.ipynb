```python
import matplotlib.pyplot as plt
# 한글 설정
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic' # Window용
# matplotlib.rcParams['font.family'] = 'AppleGothic' # Mac 용
matplotlib.rcParams['font.size'] = 15 # 글자크기
matplotlib.rcParams['axes.unicode_minus'] = False # 그래프 마이너스까지 나타낼 수 있게
```

```python
### x, y축 데이터 (리스트,튜플,DataFrame - 리스트 타입, Series)
x = [1,2,3]
y = [2,4,8] # SM 매출액
y2 = [5,1,3] # LG 매출액
y3 = [1,2,5] # SK 매출액

plt.figure(figsize=(10,5))
plt.plot(x,y,label='SM 매출액',marker='o') # 
plt.plot(x,y2,label='LG 매출액',marker='v')
plt.plot(x,y3,label='SK 매출액',marker='x')

# for idx,txt in enumerate(y):
#     plt.text(x[idx],y[idx]+0.2,txt,ha='center')
for i in range(3):
    plt.text(x[i],y[i])

plt.title(' 기업별 매출액 ')
plt.legend(ncol=3) # 범례출력 기본값: 세로출력
plt.show()
```
