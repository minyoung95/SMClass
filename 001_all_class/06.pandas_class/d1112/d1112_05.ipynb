```python
import matplotlib.pyplot as plt
# 한글 설정
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic' # Window용
# matplotlib.rcParams['font.family'] = 'AppleGothic' # Mac 용
matplotlib.rcParams['font.size'] = 15 # 글자크기
matplotlib.rcParams['axes.unicode_minus'] = False # 그래프 마이너스까지 나타낼 수 있게
```

```python
import pandas as pd
df = pd.read_excel('score.xlsx',index_col='지원번호')
```

```python
# x축 지원번호, y축 국어
# 선색상: red, marker색상: yellow, marker테두리: blue, 선타입: --
x = df.index
k = df['국어']
m = df['수학']
e = df['영어']

### 그래프 전체 크기 figure=(x,y) : x축 y축 크기 조절, facecolor : 배경색
plt.figure(figsize=(20,5),facecolor='yellow') # dpi : 그래프해상도 (높을수록 이미지 선명해짐)
plt.title('[ 학생성적그래프 ]',size=20)
plt.plot(x,k,label='국어점수',color='red',marker='o',mfc='yellow',mec='blue',ls='--')
plt.plot(x,m,'bo--',label='수학점수',mfc='yellow',mec='blue')
plt.plot(x,e,'go--',label='영어점수',mfc='yellow',mec='blue')

plt.legend(loc='lower right')
### savefig : 파일저장
# plt.savefig('graph2.png',dpi=300)
plt.show()
```
