```python
import matplotlib.pyplot as plt
# 한글 설정
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic' # Window용
# matplotlib.rcParams['font.family'] = 'AppleGothic' # Mac 용
matplotlib.rcParams['font.size'] = 15 # 글자크기
matplotlib.rcParams['axes.unicode_minus'] = False # 그래프 마이너스까지 나타낼 수 있게
```

```python
label = ['홍길동','유관순','강감찬']
values = [190,187,184]

plt.figure(figsize=(7,5))
plt.title('막대그래프')
colors = ['r','g','b'] # 바 색상 여러개로 처리
# 막대그래프 bar
# plt.bar(label,values,color='r')
plt.bar(label,values,color=colors,width=0.2,label='학생키') # width : 바 크기
plt.xticks(label,['1번','2번','3번'],rotation=45) # label,[] : 텍스트 수정, rotation 기울기
plt.yticks([0,50,100,150,200,250])
plt.legend(loc='upper center')
plt.show()
```

```python
colors = ['r','g','b']
bar = plt.bar(label,values,width=0.5,color=colors)
# 막대그래프 무늬
bar[0].set_hatch('/')
bar[1].set_hatch('x')
bar[2].set_hatch('..')
# plt.yticks([0,100,150,200]) # 0부터 범위지정

plt.ylim(175,199) # y축 범위지정
plt.show()
```

```python
label = ['홍길동','유관순','강감찬']
values = [190,187,184]
colors = ['r','g','b']
# barh : 막대그래프 눕히기
plt.barh(label,values,color=colors,height=0.5)
# plt.yticks([0,100,150,200]) # 0부터 범위지정
plt.xlim([175,190])
plt.show()
```
