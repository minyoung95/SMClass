```python
import pandas as pd
import numpy as np
df = pd.read_csv('chipotle.tsv',sep='\t')
df
```

```python
## 메뉴 별 주문수량 출력
df.groupby('item_name')['quantity'].sum()

# ## 메뉴 별 총 주문량 출력

## 주문당 평균 계산 금액 출력
# df['item_price'] = df['item_price'].str[1:].astype(float)
df.groupby('order_id')['item_price'].sum().mean()
## 아이템별 가격 금액 출력
df.groupby('item_name')['item_price'].unique()
## 주문별, 높은 가격 순으로 주문 갯수 출력
df.groupby('order_id')['quantity'].sum().sort_values(ascending=False)
## 주문당 평균 계산금액 출력
df.groupby('order_id')['item_price'].mean()
```
