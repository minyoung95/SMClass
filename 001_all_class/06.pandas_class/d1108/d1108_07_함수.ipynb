```python
import pandas as pd
df = pd.read_excel('score.xlsx',index_col='지원번호')
df
```

함수 선언 및 호출 : apply()

```python
def add_cm(height):
    return str(height) + 'cm'
```

```python
## apply 함수 호출
df['키'] = df['키'].apply(add_cm)
df
```

```python
df.info() # 키 컬럼 오브젝트로 바뀜
```

```python
df['SW특기'] = df['SW특기'].str.lower()
```

```python
def change(lang): # lang타입 : str
    if pd.notnull(lang):
        return lang.upper()
    else:
        return 'PYTHON'
```

```python
df['SW특기'].apply(change)
```
