```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
# 한글 설정
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic' # Window용
# matplotlib.rcParams['font.family'] = 'AppleGothic' # Mac 용
matplotlib.rcParams['font.size'] = 15 # 글자크기
matplotlib.rcParams['axes.unicode_minus'] = False # 그래프 마이너스까지 나타낼 수 있게

df = pd.read_excel('score.xlsx',index_col='지원번호')
```

```python
df['학년'] = [3,3,2,1,1,3,2,2]
```

산점도그래프

```python
sizes = df['학년']*500
plt.title('산점도그래프')
# s: 산점도 크기변경, c: 산점도 표시 컬럼, cmap:색상표 사용하여 색 입히기
plt.scatter(df['영어'],df['수학'],s=sizes,c=df['학년'],cmap='plasma',alpha=0.5)
plt.xlabel('영어')
plt.ylabel('수학')
# ticks: 색상바 범위, shrink: 색상바 크기, orientation : 색상바 위치(horizontal : 아래)
plt.colorbar(ticks=[1,2,3],label='학년',shrink= 0.5,orientation= 'horizontal') # 색상
```
