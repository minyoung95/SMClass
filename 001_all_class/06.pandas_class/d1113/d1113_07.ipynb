```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
# 한글 설정
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic' # Window용
# matplotlib.rcParams['font.family'] = 'AppleGothic' # Mac 용
matplotlib.rcParams['font.size'] = 15 # 글자크기
matplotlib.rcParams['axes.unicode_minus'] = False # 그래프 마이너스까지 나타낼 수 있게

# df = pd.read_excel('score.xlsx',index_col='지원번호')
```

```python
data = {
   '영화' : ['명량', '극한직업', '신과함께-죄와 벌', '국제시장', '괴물', '도둑들', '7번방의 선물', '암살'],
   '개봉 연도' : [2014, 2015, 2017, 2014, 2015, 2017, 2013, 2015],
   '관객 수' : [1761, 1626, 1441, 1426, 1301, 1298, 1281, 1270], # (단위 : 만 명)
   '평점' : [8.88, 9.20, 8.73, 9.16, 8.62, 7.64, 8.83, 9.10]
}

df = pd.DataFrame(data)
df
```

```python
## 영화 관객 수를 막대그래프로 표현
# 평균 막대그래프 추가, 평균 선 긋기 (이름: 평균, 년도 2024, 관객수: 평균관객수, 평점 : 평균평점)
df.loc[8] = ['평균',2024,df['관객 수'].mean(),df['평점'].mean()]

values = df['영화']
labels = df['관객 수']
data = df['관객 수'].mean()
plt.figure(figsize=(10,7))
s_bar = plt.bar(values,labels,width=0.5)
s_bar[len(df.index)-1].set_color('red')
plt.plot(df.index,[data]*9, ls='--', color='red')
plt.text(df['영화'][0],df['관객 수'][0],int(df['관객 수'][0]),ha='center')
plt.xticks(rotation = 45)
plt.grid(axis= 'y',ls='--',alpha=0.5)
plt.ylim(1200,2000)

plt.show()
```
