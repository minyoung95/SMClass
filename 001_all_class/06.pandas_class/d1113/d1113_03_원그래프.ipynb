```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
# 한글 설정
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic' # Window용
# matplotlib.rcParams['font.family'] = 'AppleGothic' # Mac 용
matplotlib.rcParams['font.size'] = 15 # 글자크기
matplotlib.rcParams['axes.unicode_minus'] = False # 그래프 마이너스까지 나타낼 수 있게
```

원 그래프

```python
# values = [30,25,20,13,10,2] # 100퍼 맞춤
values = [1]*6 # 비율 6등분
labels = ['python','java','javascript','oracle','html','etc']
explode = [0.2,0,0,0.1,0,0]
colors = ['plum','beige','skyblue','bisque','khaki','ivory']
plt.figure(figsize=(8,6))
plt.title('프로그램 사용률')
# explode : 원그래프 분리, counterclock : 시계반대방향(False : 시계방향), 
# startangle=90 : 12시부터 시작 (3시 0도 기준), autopct : 데이터값 출력 %f%%
plt.pie(values,labels=labels,explode=explode,autopct='%.1f%%',counterclock=False,startangle=90,colors=colors)
plt.legend(loc=(1,0.2),title='언어별 선호도')
plt.show()
```
