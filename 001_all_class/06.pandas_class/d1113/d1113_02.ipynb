```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
# 한글 설정
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic' # Window용
# matplotlib.rcParams['font.family'] = 'AppleGothic' # Mac 용
matplotlib.rcParams['font.size'] = 15 # 글자크기
matplotlib.rcParams['axes.unicode_minus'] = False # 그래프 마이너스까지 나타낼 수 있게
```

```python
df = pd.read_excel('score.xlsx',index_col='지원번호')
```

```python
def cm_func(x):
    return str(x)+'cm'
```

```python
## df['키'](배열타입) : 모든 것을 함수적용 - apply 
df['키'] = df['키'].apply(cm_func)
df['키']
```

```python
## 50점 아래 있는 값 : 50으로 리턴해서 출력
def k_func(x):
    if x>=50: return x
    return 50

# int,str,float 같은 단일값은 apply X, 함수명(데이터)로 호출
for i in range(8):
    print(k_func(df['국어'][i])) # df['국어']의 단일값이므로 apply X
```

```python
def func(x):
    return int(x/2)

plt.bar(df['이름'],df['국어'])
plt.bar(df['이름'],df['영어'],bottom=df['국어'])
plt.bar(df['이름'],df['수학'],bottom=df['국어']+df['영어'])

for i in range(8):
    plt.text(df['이름'].iloc[i],func(df['국어'].iloc[i]),df['국어'].iloc[i], ha='center')
    plt.text(df['이름'].iloc[i],df['국어'].iloc[i]+func(df['영어'].iloc[i]),df['영어'].iloc[i], ha='center')
    plt.text(df['이름'].iloc[i],df['국어'].iloc[i]+df['영어'].iloc[i]+func(df['수학'].iloc[i]),df['수학'].iloc[i], ha='center')
```
