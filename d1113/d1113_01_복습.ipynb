```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
# 한글 설정
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic' # Window용
# matplotlib.rcParams['font.family'] = 'AppleGothic' # Mac 용
matplotlib.rcParams['font.size'] = 15 # 글자크기
matplotlib.rcParams['axes.unicode_minus'] = False # 그래프 마이너스까지 나타낼 수 있게
```

```python
df = pd.read_excel('score.xlsx',index_col='지원번호')
df
```

```python
plt.title('학생성적그래프')
plt.plot(df.index,df['국어'], label='국어')
plt.plot(df.index,df['영어'], label='영어')
plt.plot(df.index,df['수학'], label='수학')

for i in range(8):
    plt.text(i,df['국어'].iloc[i]+0.6,df['국어'].iloc[i],ha='center')
    plt.text(i,df['영어'].iloc[i]+0.6,df['영어'].iloc[i],ha='center')
    plt.text(i,df['수학'].iloc[i]+0.6,df['수학'].iloc[i],ha='center')

plt.legend()
plt.show()
```

누적막대그래프

```python
def n_func(x):
    if x-2 > 0: return x
    else: return 0

n_list = [10,5,-1,6,7]
for i in n_list:
    print('데이터 값 : ', n_func(i-2))
```

```python
def func(x):
    if x-20<0: return 0
    else: return x-30
```

```python
plt.figure(figsize=(8,4))
plt.title('학생성적그래프')
plt.bar(df['이름'],df['국어'],width=0.6,label='국어')
plt.bar(df['이름'],df['영어'],width=0.6,bottom=df['국어'], label='영어')
plt.bar(df['이름'],df['수학'],width=0.6,bottom=df['국어']+df['영어'], label='수학')
plt.xticks(rotation = 45)

for i in range(8):
    plt.text(df['이름'].iloc[i],func(df['국어'].iloc[i]),df['국어'].iloc[i],ha='center')
    plt.text(df['이름'].iloc[i],df['국어'].iloc[i]+df['영어'].iloc[i]-10,df['영어'].iloc[i],ha='center')
    plt.text(df['이름'].iloc[i],df['국어'].iloc[i]+df['영어'].iloc[i]+df['수학'].iloc[i],df['수학'].iloc[i],ha='center')

plt.legend()
plt.show()
```
