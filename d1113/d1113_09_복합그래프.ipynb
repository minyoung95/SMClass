```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
# 한글 설정
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic' # Window용
# matplotlib.rcParams['font.family'] = 'AppleGothic' # Mac 용
matplotlib.rcParams['font.size'] = 15 # 글자크기
matplotlib.rcParams['axes.unicode_minus'] = False # 그래프 마이너스까지 나타낼 수 있게

df = pd.read_excel('score.xlsx',index_col='지원번호')
```

복합그래프

```python
# 2*2 4개 그래프 혼합
fig,axs = plt.subplots(2,2,figsize=(15,13))
fig.suptitle('4개 복합 그래프') # 메인 타이틀

# 4개 그래프 : 1번째 axs[0,0],[0,1],[1,0],[1,1]
# 1번째 : bar
axs[0,0].bar(df['이름'],df['국어'],label='국어')
axs[0,0].set_title('1.국어점수') # 서브 타이틀
axs[0,0].set(xlabel='이름',ylabel='점수') # x,y축
axs[0,0].set_facecolor('lightyellow') # 배경
axs[0,0].legend(loc='upper center')
# 2번째 : plot
axs[0,1].plot(df['이름'],df['영어'],label='영어')
axs[0,1].plot(df['이름'],df['수학'],label='수학')
axs[0,1].set_title('2.영어,수학점수')
axs[0,1].set(xlabel='이름',ylabel='점수')
axs[0,1].legend()
# 3번째 : barh
axs[1,0].barh(df['이름'],df['키'],label='키')
axs[1,0].set_xlim(150,210)
axs[1,0].set_title('3. 키')
for i in range(8):
    axs[1,0].text(df['키'].iloc[i]+3,df['이름'].iloc[i],df['키'].iloc[i],ha='center')
# 4번째 : plot
axs[1,1].plot(df['이름'],df['사회'],color='g',alpha=0.5)
axs[1,1].set_title('4.사회점수')
axs[1,1].set(xlabel='이름',ylabel='점수')
for i in range(8):
    axs[1,1].text(df['이름'].iloc[i],df['사회'].iloc[i]+2,int(df['사회'].iloc[i]),ha='center')
plt.show()
```
