```python
import pandas as pd
```

```python
# 상단 3개 삭제 후 가져오기
# index 행정기관을 지정하여 df로 변환해서 출력
# 위 5개만 출력

df = pd.read_excel('연령별인구현황01.xlsx',skiprows=3,index_col='행정기관',usecols='B,E:Y') # skiprows=x : 상단 x줄 삭제, usecols='X,Y:Z' : 원하는 컬럼을 가져옴 (X와 Y~Z사이 가져옴)
df.head() # 상단 5개
```

```python
df[0:1]
df.iloc[0] # row을 보는 명령어
df.loc['전국  '] # row을 보는 명령어
df.iloc[0] = df.iloc[0].str.replace(',','').astype(int) # str : 문자형으로 정의 후 // replace (문자형만 가능) // astype : 타입변환
# df.info()
df
```
