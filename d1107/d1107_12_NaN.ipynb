```python
import pandas as pd
import numpy as np
df = pd.read_excel('score.xlsx',index_col='지원번호')
df
```

fillna() : NaN 데이터 값을 해당 값으로 채워줌

dropna() : Nan 데이터를 삭제 // axis(축) : index - row방향, columns - columns 방향 // how : any - NaN값이 하나라도 있을 경우, all - 모든 값이 NaN일 경우

np.nan : NaN 값으로 만들기

```python
### 결측치 (NaN데이터)
# NaN데이터 채우기 : fillna()

df.fillna('없음',inplace=True)
df
```

```python
## 학교 컬럼에 NaN 데이터 입력 : np.nan
df['학교'] = np.nan
df.fillna('없음') # inplace= True, 모든 NaN값에 해당데이터 입력
df['SW특기'].fillna('없음',inplace=True) # SW특기 컬럼의 NaN에만 해당 데이터 입력
df
```

```python
df = pd.read_excel('score.xlsx',index_col='지원번호')
df
```

```python
# NaN 데이터 삭제 : dropna()
df.dropna() # NaN 데이터가 있는 row를 모두 삭제
```

```python
# axis : 축 정하기, how : default = any (NaN값이 한개라도 있을 때 삭제), all (전부 NaN일때 삭제)
df.dropna(axis='index',how='any') # index 방향(row)으로 
```

```python
df = pd.read_excel('score.xlsx',index_col='지원번호')
df
```

```python
df.dropna(axis='columns',how='all') # columns 방향 (세로), 전체가 NaN일때 삭제
df['학교'] = np.nan # 학교 컬럼 전체를 NaN값으로 만들기
df.dropna(axis='columns',how='all')
```
