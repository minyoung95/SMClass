```python
import matplotlib.pyplot as plt
# 한글 설정
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic' # Window용
# matplotlib.rcParams['font.family'] = 'AppleGothic' # Mac 용
matplotlib.rcParams['font.size'] = 15 # 글자크기
matplotlib.rcParams['axes.unicode_minus'] = False # 그래프 마이너스까지 나타낼 수 있게
```

```python
### x, y축 데이터 (리스트,튜플,DataFrame - 리스트 타입, Series)
x = [1,2,3]
y = [2,4,8]
yy = [1,8,3]
## linewidth : 선 두께, alpha : 투명도, linestyle(ls) : 선모양 (:,-,--,-.)
plt.plot(x,y, label='인원증가',linewidth=5,alpha=0.8,ls=':')
## marker : 꺾이는 지점 표시 (o,x,v),markersize(ms),markeredgecolor(mec:마커 테두리선 색),markerfacecolor(mfc:마커 내부 색)
# plt.plot(x,yy,'bv:',label='비용증가',marker='o',ms=10,mec='green')
## b: 선색상, v: 마커표시,: 선모양
plt.plot(x,yy,'bx:',label='비용증가',marker='o',ms=10,mec='green')
### 상단 타이틀
plt.title('라인그래프 Line',size=20) # size: 글자크기

### 범례 : 그래프 선 설명문구
# plt.legend(loc='upper right') # loc 설명문구 위치 (upper,center,lower/left,center,right)
plt.legend(loc=(0.35,0.05))
```
