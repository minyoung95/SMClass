```python
import pandas as pd
import matplotlib.pyplot as plt
# 한글 설정
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic' # Window용
# matplotlib.rcParams['font.family'] = 'AppleGothic' # Mac 용
matplotlib.rcParams['font.size'] = 15 # 글자크기
matplotlib.rcParams['axes.unicode_minus'] = False # 그래프 마이너스까지 나타낼 수 있게
```

```python
df = pd.read_excel('score.xlsx')

plt.plot(df['이름'],df['키'])
plt.xticks(rotation=45)
plt.grid(axis='y',color='purple',ls='--',linewidth=2,alpha=0.2) # 그래프에 안내선 긋기 (axis=x,y (x,y축 기준))
plt.grid(axis='x',alpha=0.2)

# 그래프 값 추가
for i in range(8):
    plt.text(df['이름'][i],df['키'][i]+1,df['키'][i],ha='center')
plt.show()
```
