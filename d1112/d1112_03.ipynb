```python
import matplotlib.pyplot as plt
# 한글 설정
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic' # Window용
# matplotlib.rcParams['font.family'] = 'AppleGothic' # Mac 용
matplotlib.rcParams['font.size'] = 15 # 글자크기
matplotlib.rcParams['axes.unicode_minus'] = False # 그래프 마이너스까지 나타낼 수 있게
```

```python
# x, y축 데이터 (리스트,튜플,DataFrame - 리스트 타입, Series)
x = [1,2,3]
y = [2,4,8]

plt.plot(x,y)
## 상단 타이틀
plt.title('라인그래프 Line',size=20) # size: 글자크기
## 라벨 (color : color picker)
plt.xlabel('X축 날짜',color='#00ff00',loc='right') # 그래프 x축 설명,color,loc(글자 정렬 : left,center,right)
plt.ylabel('Y축 인원',color='blue',loc='top') # 그래프 y축 설명,color,loc(글자 정렬 : top,center,bottom)

## 그래프 눈금단위
plt.xticks([1,2,3])
plt.yticks([2,3,4,8,9,10])



plt.show()
```
